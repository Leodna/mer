```python
import numpy as np
import matplotlib.pyplot as plt

# Simulamos la carga de 20 imágenes como matrices aleatorias
imagenes_procesadas = [np.random.rand(100, 100) for _ in range(20)]

# Convertimos la lista de imágenes en un array de numpy
imagenes_array = np.array(imagenes_procesadas)

# Seleccionamos 10 imágenes al azar sin reemplazo
indices_aleatorios = np.random.choice(imagenes_array.shape[0], 10, replace=False)
imagenes_aleatorias = imagenes_array[indices_aleatorios]

# Graficamos las 10 imágenes
fig, axes = plt.subplots(2, 5, figsize=(15, 6))

for i, ax in enumerate(axes.flat):
    ax.imshow(imagenes_aleatorias[i], cmap='gray')
    ax.axis('off')

plt.show()
```
